# In this project, I will look at:-
    
- The gender gap in the tech industry and why it exists.
- Data visualization of the gender gap in tech.
- What we can do about it.

```python
# imports all the necessary libraries..
import seaborn as sns
color = sns.color_palette()
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
data = pd.read_csv('multiple_responses.csv')
```

```python
data.head(5)
```

```python
# let's only pick out columns that are useful that we need.
data = data[['Q1','Q2','Q3','Q4','Q5','Q10']]
data.columns = ['age','gender','country','level_education','job_title','salary']
data = data.drop(data.index[0])
data.head()
```

```python
data.info()
```

# Visualize age distribution

```python
sns.set_style('darkgrid')
plt.figure(figsize = (10, 5))
sns.countplot(x = 'age', data =data)
```

It could be observed that most survey respondents are fairly young, around 20 to 30 years of age.

```python
sns.set_style('darkgrid')
plt.figure(figsize = (10, 5))
sns.countplot(x = 'gender', data =data)
```

Most of the respondents are males which carries the higher proportion.

# Age distribution of all respondent alongside with gender

```python
sns.set_style('darkgrid')
plt.figure(figsize = (10, 5))
sns.countplot(x = 'age', hue = 'gender', data = data)
```

# Education level

```python
sns.set_style('darkgrid')
plt.figure(figsize = (15, 5))
sns.countplot(x = 'level_education', data =data)
plt.xticks(rotation = 60, horizontalalignment = 'right');
```

As it could be observed, most of the respondents have a Master’s, or at least a Bachelor’s degree.

#Education level alongside age

```python
sns.set_style('darkgrid')
plt.figure(figsize = (15, 5))
sns.countplot(x = 'level_education', hue = 'age', data =data)
plt.xticks(rotation = 60, horizontalalignment = 'right');
```

Education level alongside gender

```python
sns.set_style('darkgrid')
plt.figure(figsize = (15, 5))
sns.countplot(x = 'level_education', hue = 'gender' ,data =data)
plt.xticks(rotation = 60);
```

The education levels of both male and female respondents show similar patterns — female respondents are just as well educated as their male counterparts.

# Salary distribution

```python
sns.set_style('darkgrid')
plt.figure(figsize = (15, 5))
plot = sns.countplot(x = 'salary', data =data)
plt.xticks(rotation = 60, horizontalalignment = 'right');
```

salary vs gender

```python
sns.set_style('darkgrid')
plt.figure(figsize = (15, 5))
sns.countplot(x = 'salary', hue = 'gender' ,data =data)
plt.xticks(rotation = 60);
```

Job distribution

```python
sns.set_style('darkgrid')
plt.figure(figsize = (15, 5))
plot = sns.countplot(x = 'job_title', data =data)
plt.xticks(rotation = 60, horizontalalignment = 'right');
```

Most respondents seem to be students or data scientists. There are only a infinitesimal numbers of statisticians and database engineers.

Job distribution against gender

```python
sns.set_style('darkgrid')
plt.figure(figsize = (15, 5))
plot = sns.countplot(x = 'job_title', hue = 'gender', data =data)
plt.xticks(rotation = 60, horizontalalignment = 'right');
```

# Conclusion

To conclude, gender disparity in the tech industry is real and women are severely underrepresented in this field.

So what are the solution(s) to this problem?

- Girls should be taught that it’s OK to make mistakes and that trial and error is part of life.
- Special programs can be conducted for girls by women who are already in the industry, to encourage girls to join the field. 
- Girls need to be exposed to the massive changes that can be brought about by pursuing careers in the tech sector.
